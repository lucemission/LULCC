## Melihat banyak data koordinat latitude dan longitude

```python
lats = []
lons = []
# os.listdir('./lu')
for filename in os.listdir('./data/1500-2004/lu'):
    lat = re.search('lat(.+?)lon', filename).group(1)
    lon = re.search('lon(.+?)\.lu', filename).group(1)
    if lat not in lats:
        lats.append(lat)
    if lon not in lons:
        lons.append(lon)
    
lats = sorted([float(i) for i in lats])
lons = sorted([float(i) for i in lons])

print(f"Banyak koordinat latitude: {len(lats)}")
print(f"Banyak koordinat longitude: {len(lons)}")
```

## Melihat metadata dari file-file di dalam folder updated_states

```python
# Variabel untuk menampung nilai-nilai metadata
cols = []    # banyak koordinat longitude
rows = []    # banyak koordinat latitude
xll = []     #
yll = []     # 
sizes = []   #
nodatas = [] #
counter = 0

# os.listdir('./lu')
for file in os.listdir('./updated_states'):
    with open(f"./updated_states/{file}", 'r') as readfile:
        text = readfile.read()
        col = re.search('ncols\s+(.+?)\s', text).group(1)
        row = re.search('nrows\s+(.+?)\s', text).group(1)
        x = re.search('xllcorner\s+(.+?)\s', text).group(1)
        y = re.search('yllcorner\s+(.+?)\s', text).group(1)
        size = re.search('cellsize\s+(.+?)\s', text).group(1)
        nodata = re.search('NODATA_value\s+(.+?)\s', text).group(1)
        
        if col not in cols:
            cols.append(col)
        if row not in rows:
            rows.append(row)
        if x not in xll:
            xll.append(x)
        if y not in yll:
            yll.append(y)
        if size not in sizes:
            sizes.append(size)
        if nodata not in nodatas:
            nodatas.append(nodata)
            
        counter+=1
```

```python
print(cols)
print(rows)
print(xll)
print(yll)
print(sizes)
print(nodatas)
```
